```python
import requests
import pandas as pd
import json
```

```python
api_url = "https://www.alphavantage.co/query"
api_key = "9NAVJMDOR80YHPT0"
params = {
    "function": "TIME_SERIES_INTRADAY",
    "symbol": "IBM",
    "interval": "5min",
    "apikey": api_key
}

response = requests.get(api_url,params=params)
data = response.json()

print(json.dumps(data, indent=4))  # should show 'Meta Data' and 'Time Series (5min)'

```

```python
data = data['Time Series (5min)']
```

```python
data
```

```python
data_temp = []
for key,value in data.items():
    data_temp.append(
        {
            'tmie_stamp':key,
            'open':value['1. open'],
            'high':value['2. high'],
            'low':value['3. low'],
            'close':value['4. close'],
            'volume':value['5. volume']
        }
    )
data = pd.DataFrame(data_temp)
```
